```python
import pandas as pd
import numpy as np
```

```python
data = {

 'Student_ID': [101, 102, 103, 104, 105, 106, 107],
 'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eva', 'Frank', 'Grace'],
 'Gender': ['F', 'M', 'M', 'F', 'F', 'M', 'F'],
 'Math': [85, 67, 90, 78, 92, 58, 73],
 'Science': [88, 72, 94, 74, 91, 60, 77],
 'English': [90, 65, 88, 82, 85, 70, 79],
 'Attendance (%)': [96, 82, 89, 94, 99, 76, 88]
 
}
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df[ df ['Attendance (%)'] < 85] ['Name']
```

```python
df['Total'] = df['Math']+df['Science']+df['English']
df[['Name', 'Total']]
```

```python
df['Average'] = df['Total']/3
df[['Name', 'Average']]
```

```python
df.rename(columns={'Attendance (%)': 'Attendance'},inplace=True)
df.head()
```

```python
df.sort_values(by='Total', ascending=False)
```

```python
df.groupby('Gender')[['Math','Science','English','Total','Average']].mean()
```

```python
max_english = df['English'].max()
df[df['English']==max_english][['Name','English']]
```

```python
df['Grade'] = df['Average'].apply(lambda avg: 'A' if avg >= 85 else ('B' if avg >= 70 else 'C'))
df[['Name','Grade']]
```

```python
df['Grade'].value_counts()
```

```python
df.loc[0, 'Math'] = np.nan
df.loc[1, 'Attendance'] = np.nan
df.head()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```
