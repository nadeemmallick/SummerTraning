<h2>Student Scores Analysis</h2>

#### 1.Import Libr..

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

#### 2.Data set

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ivy", "Jack"],
    "Math": [85, 78, 90, 88, 76, 95, 89, 67, 72, 80],
    "Science": [82, 75, 92, 85, 80, 97, 88, 65, 70, 77],
    "English": [78, 83, 85, 80, 75, 90, 86, 72, 74, 79],
    "Gender": ["F", "M", "M", "M", "F", "M", "F", "F", "F", "M"]
}

df = pd.DataFrame(data)
df
```

#### 3. Bar chart of Math scores.

```python
plt.figure(figsize=(6, 3))
sns.barplot(x="Name", y="Math", data=df,palette="viridis")
plt.title("Math Scores of All Students")
plt.xticks(rotation=35)
plt.tight_layout()
plt.show()
```

#### 4. Science scores grouped by gender

```python
plt.figure(figsize=(7, 4))
sns.boxplot(x="Gender", y="Science", data=df, palette="coolwarm")
plt.title("Science Scores by Gender")
plt.tight_layout()
plt.show()
```

#### 5. Line chart of average scores per subject

```python
average_scores = {
    'Subject': ['Math', 'Science', 'English'],
    'Average': [df['Math'].mean(), df['Science'].mean(), df['English'].mean()]
}
average_df = pd.DataFrame(average_scores)
average_df
```

```python
plt.figure(figsize=(7, 4))
sns.lineplot(x="Subject", y="Average", data=average_df, marker='o', color='black')
plt.title("Average Scores per Subject")
plt.ylim(0, 100)
plt.tight_layout()
plt.show()
```

#### 6. Seaborn Pairplot() Relationship b/w (M,S,E)

```python
sns.pairplot(df[["Math", "Science", "English"]])
plt.suptitle("Pairplot of Subject Scores", y=1.03)
plt.show()
```

#### 7. Correlation Heatmap

```python
plt.figure(figsize=(6, 4))
sns.heatmap(df[["Math", "Science", "English"]].corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
